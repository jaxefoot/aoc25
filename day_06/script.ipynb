# Part 1

```python
import re
import pandas as pd

def prepare_line(line):
    line = line.strip()
    line_content = re.split(r"[ \t]+", line)
    return list(line_content)

with open("data.txt") as f:
    lines = f.readlines()
    lines = list(map(prepare_line, lines))
    df = pd.DataFrame(lines)
```

```python
total = 0
for series_name, series in df.items():
    result = 0
    operator = series.iloc[-1]
    values = series[0:-1]
    values = values.map(lambda x: int(x))
    if operator == "+":
        result = values.sum()
    if operator == "*":
        result = values.prod()
    total += result

total
```

# Part 2

```python
import numpy as np

with open("data.txt") as f:
    lines = f.readlines()
    lines = list(map(lambda x: x[0:-1], lines))
    lines = list(map(lambda x: list(x), lines))
    df = pd.DataFrame(lines)
    df = df.fillna("")
```

```python

row_idx, col_idx = np.where(df.isin(["+", "*"]))
```

```python
column_sections = []

# get inclusive start and end point of each section
for col_idx_idx in range(len(col_idx)):
    start = col_idx[col_idx_idx]
    stop = df.shape[1]-1
    if(col_idx_idx < len(col_idx) - 1):
        stop = col_idx[col_idx_idx+1]-2
    column_sections.append((start, stop))

total = 0
values_df = df.head(-1)

# go through each section and calculate the result
for column_area in column_sections:
    operator = df.iloc[df.shape[0]-1, column_area[0]]    
    result = 0 if operator == "+" else 1

    for column_idx in range(column_area[0], column_area[1]+1):
        number = int(values_df[column_idx].sum().strip())
        if operator == "*":
            result *= number
        else:
            result += number
    total += result

total
```

```python
import re
import pandas as pd

# part 1
def prepare_line(line):
    line = line.strip()
    line_content = re.split(r"[ \t]+", line)
    return list(line_content)

with open("data.txt") as f:
    lines = f.readlines()
    lines = list(map(prepare_line, lines))
    df = pd.DataFrame(lines)

df
```

```python
# part 1
grand_total = 0
for series_name, series in df.items():
    result = 0
    operator = series.iloc[-1]
    values = series[0:-1]
    values = values.map(lambda x: int(x))
    if operator == "+":
        result = values.sum()
    if operator == "*":
        result = values.prod()
    grand_total += result

grand_total
```

# Part 2

```python
import numpy as np

with open("data.txt") as f:
    lines = f.readlines()
    lines = list(map(lambda x: x[0:-1], lines))
    lines = list(map(lambda x: list(x), lines))
    df = pd.DataFrame(lines)
    df = df.fillna("")
```

```python

row_idx, col_idx = np.where(df.isin(["+", "*"]))
```

```python
column_areas = []
for col_op_idx in range(len(col_idx)):
    start = col_idx[col_op_idx]
    stop = df.shape[1]-1
    if(col_op_idx < len(col_idx) - 1):
        stop = col_idx[col_op_idx+1]-2
    column_areas.append((start, stop))

total = 0
values_df = df.head(-1)
for column_area in column_areas:
    operator = df.iloc[df.shape[0]-1, column_area[0]]
    result = 0
    if operator == "*":
        result = 1

    for column_idx in range(column_area[0], column_area[1]+1):
        number = int(values_df[column_idx].sum().strip())
        if operator == "*":
            result *= number
        else:
            result += number
    total += result

total
```

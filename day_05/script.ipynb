```python
def create_range(range_string):
    start, stop = range_string.split("-")
    return (int(start), int(stop))
    
    
with open("data.txt") as f:
    file = f.read()
    ranges, ingredients = file.split("\n\n")
    ranges = ranges.split("\n")
    ranges = list(map(create_range, ranges))
    
    ingredients = ingredients.split("\n")
    ingredients = list(map(lambda x: int(x), ingredients))
```

```python
# Part 1
fresh_ingredient_counter = 0

for ingredient in ingredients:
    for range in ranges:
        if ingredient >= range[0] and ingredient<= range[1]:
            fresh_ingredient_counter += 1
            break
fresh_ingredient_counter
```

```python
# Part 2
def get_new_ids_of_added_range(ranges, new_range):
    updated_ranges = [new_range]
    for range in ranges:
        new_updated_ranges = []
        for updated_range in updated_ranges:
            # updated range is included in range
            if updated_range[0] >= range[0] and updated_range[1] <= range[1]:
                break
            
            # left overlap or on the left
            if updated_range[0] < range[0]:
                new_updated_ranges.append((updated_range[0], min(updated_range[1], range[0]-1)))
            # right overlap or on the right
            if updated_range[1] > range[1]:
                new_updated_ranges.append((max(range[1]+1, updated_range[0]), updated_range[1]))
        
        updated_ranges = new_updated_ranges
    return updated_ranges

accepted_ranges = []
for range in ranges:
    addable_id_ranges = get_new_ids_of_added_range(accepted_ranges, range)
    accepted_ranges.extend(addable_id_ranges)

number_of_ids = map(lambda x: x[1]-x[0]+ 1, accepted_ranges)
sum(number_of_ids)
```

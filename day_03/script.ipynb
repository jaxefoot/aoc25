```python
with open("data.txt") as f:
    banks = f.readlines()
    banks = list(map(lambda x: x.strip(), banks))
```

# Solution Task 1

```python
joltage = 0
for bank in banks:
    batteries = list(bank)
    
    #find first number
    highest_tenner = max(batteries[:-1])
    tenner_index = batteries[:-1].index(highest_tenner)
    
    highest_single = max(batteries[tenner_index+1:])
    bat_joltage = int(highest_tenner+highest_single)
    joltage += bat_joltage
joltage
    
```

# Solution Task 2

```python
def get_total_joltage(banks, n):
    joltage = 0
    for bank in banks:
        batteries = list(bank)
        start = 0
        turned_batteries = []
        for n_bat in range(0, n):
            stop = (n-n_bat-1)*(-1)
            if stop == 0:
                stop = len(batteries)
            available_batteries = batteries[start:stop]
            turned_battery = max(available_batteries)
            start = batteries.index(turned_battery, start, stop)+1
            turned_batteries.append(turned_battery)
        bat_joltage = int(''.join(turned_batteries))
        joltage += bat_joltage
    return joltage
```

```python
get_total_joltage(banks=banks, n=2), get_total_joltage(banks=banks, n=12)
```

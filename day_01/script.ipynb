```python
import math

with open("data.txt") as f:
    lines = f.readlines()
```

```python
max_number = max(map(lambda x: int(x[1:]), lines))
max_number
```

```python


current_number = 50
counter = 0
bin_size = 100

for line in lines:
    direction = line[0]
    amount = int(line[1:])
    previous_number = current_number
    if direction == "L":
        current_number = previous_number-amount
    else:
        current_number = previous_number+amount

    if current_number % bin_size == 0:
        counter += 1

counter
```

```python
current_number = 50
counter = 0
for line in lines:
    direction = line[0]
    amount = int(line[1:])
    previous_number = current_number
    
    if direction == "L":
        current_number = previous_number-amount
    else:
        current_number = previous_number+amount

    # binning logic: -100 to -1 -> -1; 0 to 99 -> 0; 100 to 199 -> 1
    # 0 is always the first number of a bin
    previous_number_bin = math.floor(previous_number/bin_size)
    current_number_bin = math.floor(current_number/bin_size)

    zero_passes = abs(previous_number_bin - current_number_bin)

    # case: starts at 0 and moves left, automatically goes to new bin despite not passing 0
    if previous_number % bin_size == 0 and direction == "L":
        zero_passes -= 1
    
    # case: stays in same bin, but lands at 0
    if current_number % bin_size == 0 and direction == "L":
        zero_passes += 1

    counter += zero_passes


counter
```

```python
file_name = "data.txt"

with open(file_name) as f:
    lines = f.readlines()
    lines = list(map(lambda x: x.strip().split(","), lines))
    
red_tiles = [(int(x), int(y)) for x,y in lines]
```

```python
def get_rect_size(p1, p2):
    return (abs(p1[0]-p2[0])+1)*(abs(p1[1]-p2[1])+1)
```

```python
max_rect = 0
for idx, val in enumerate(red_tiles):
    for follow_idx in range(idx+1, len(red_tiles)):
        p1 = red_tiles[idx]
        p2 = red_tiles[follow_idx]
        rect_size = get_rect_size(p1, p2)
        if rect_size > max_rect:
            max_rect = rect_size
max_rect
```

# Part 2

```python
num_red_tiles = len(red_tiles)
green_tiles = set()

# get tiles between tiles
for idx in range(0, num_red_tiles + 1):
    current = red_tiles[idx % num_red_tiles]
    next = red_tiles[(idx + 1) % num_red_tiles]

    if current[0] == next[0]:
        x = current[0]
        ys = sorted([current[1], next[1]])
        for y in range(ys[0]+1, ys[1]):
            green_tiles.add((x, y))
    else:
        y = current[1]
        xs = sorted([current[0], next[0]])
        for x in range(xs[0]+1, xs[1]):
            green_tiles.add((x, y))
            
shape_boundaries = set(red_tiles).union(green_tiles)
```

```python
def outside_point_is_in_rect(point_a, point_b, shape_boundaries):
    xa, ya = point_a
    xb, yb = point_b
    
    xs = sorted([xa, xb])
    ys = sorted([ya, yb])
    
    result = False
    for point in shape_boundaries:
        x,y = point
        if x > xs[0] and x<xs[1] and y > ys[0] and y < ys[1]:
            result = True
            break
    return result

max_rect = 0
for idx, val in enumerate(red_tiles):
    for follow_idx in range(idx+1, len(red_tiles)):
        p1 = red_tiles[idx]
        p2 = red_tiles[follow_idx]
        rect_size = get_rect_size(p1, p2)
        if rect_size > max_rect and not outside_point_is_in_rect(p1, p2, shape_boundaries):
            max_rect = rect_size
max_rect
```

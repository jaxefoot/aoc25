```python
import textwrap

with open("data.txt") as f:
    text = f.read()
    lines = text.split(",")
```

```python
def id_is_valid_part_1(id):
    id_char_len = len(id)
    if id_char_len % 2 == 1:
        return True
    
    first_half = id[:int(id_char_len/2)]
    second_half = id[int(id_char_len/2):]
    return first_half != second_half

def id_is_valid_part_2(id):
    id_char_len = len(id)
    contains_n_repetions = False
    for number in range(1, int(id_char_len/2) + 1):
        if id_char_len % number != 0:
            continue
        chunks = textwrap.wrap(id, number)
        if len(set(chunks)) == 1:
            contains_n_repetions = True
            break
    return not contains_n_repetions

wrong_ids_1 = []
wrong_ids_2 = []  
    
for line in lines:
    id_from, id_to = line.split("-")
    id_from, id_to = int(id_from), int(id_to)

    for number in range(id_from, id_to+1):
        if not id_is_valid_part_1(str(number)):
            wrong_ids_1.append(number)
        if not id_is_valid_part_2(str(number)):
            wrong_ids_2.append(number)
        
sum(wrong_ids_1), sum(wrong_ids_2)
```

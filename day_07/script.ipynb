```python
import pandas as pd
import numpy as np
from itertools import groupby
from operator import itemgetter

with open("example.txt") as f:
    lines = f.readlines()
    lines = map(lambda x: list(x.strip()), lines)
    df = pd.DataFrame(lines)

start_row, start_col = np.where(df == "S")
starting_point = (start_row[0], start_col[0])
starting_point

splitter = np.where(df == "^")
splitter = [(row, col) for row, col in zip(splitter[0], splitter[1])]
splitter = sorted(splitter, key=itemgetter(1))
splitter_per_col = groupby(splitter, key=lambda x: x[1])
splitter_per_col = {k: list(v) for k, v in splitter_per_col}
```

# Part 1

```python
current_beams = {starting_point[1]}
num_splits = 0

for idx, data in df.iterrows():
    splitter_in_row = set(map(lambda x: x[1], filter(lambda x: x[0] == idx, splitter)))
    removed_beams = current_beams.intersection(splitter_in_row)
    current_beams = current_beams - splitter_in_row
    num_splits += len(removed_beams)
    
    starting_beams = set()
    for s in removed_beams:
        starting_beams.add(s-1)
        starting_beams.add(s+1)
    starting_beams = starting_beams
    current_beams = current_beams.union(starting_beams)

num_splits
```

# Part 2

```python
current_beams = {x: 0 for x in range(0, df.shape[1])}
current_beams[starting_point[1]] = 1
timeline_counter = 1

for idx, data in df.iterrows():
    splitter_in_row = set(map(lambda x: x[1], filter(lambda x: x[0] == idx, splitter)))
    if len(splitter_in_row) == 0:
        continue
    removed_beams = set(current_beams.keys()).intersection(splitter_in_row)
    
    starting_beams = []
    for s in removed_beams:
        beams_in_col = current_beams[s]
        timeline_counter += beams_in_col
        current_beams[s-1] += beams_in_col
        current_beams[s+1] += beams_in_col
        current_beams[s] = 0
timeline_counter
```
